```python
import numpy as np
a_load = np.load('/content/drive/MyDrive/Colab Notebooks/allimg.npy')
```

```python
import cv2
small = []
for i in range(len(a_load)):
  small.append(cv2.resize(a_load[i], (64,64)))
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(100,100))
for i in range(100):
    ax = fig.add_subplot(10, 10, i+1, xticks=[], yticks=[])
    image_plt = np.array(cv2.resize(small[i], (29, 29)))
    ax.imshow(image_plt)
    ax.set_xlabel(str(i), fontsize=20)
plt.show()
plt.close()
```

```python
for_learn = []
for i in range(len(small)):
  for_learn.append(small[i] / 255)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(100,100))
for i in range(100):
    ax = fig.add_subplot(10, 10, i+1, xticks=[], yticks=[])
    image_plt = np.array(cv2.resize(for_learn[i], (64, 64)))
    ax.imshow(image_plt)
    ax.set_xlabel(str(i), fontsize=20)
plt.show()
plt.close()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers, models
```

```python
inputs = keras.Input(shape=(64,64,1))
x = layers.Conv2D(4, 3, activation="relu")(inputs)
x = layers.MaxPooling2D(2)(x)
x = layers.Conv2D(4, 3, activation="relu")(x)
x = layers.Conv2DTranspose(4, 3, activation="relu")(x)
x = layers.UpSampling2D(2)(x)
x = layers.Conv2DTranspose(4, 3, activation="relu")(x)
outputs = x = layers.Conv2D(1, (1,1), activation="relu")(x)

model = models.Model(inputs=inputs, outputs=outputs)

model.summary()
```

```python
model.compile(
    loss=keras.losses.mse,
    optimizer=keras.optimizers.Adam()
)

history = model.fit(for_learn, for_learn, batch_size=64, epochs=2, validation_split=0.2)

test_scores = model.evaluate(for_learn, for_learn, verbose=2)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])
```
